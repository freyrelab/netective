```python
import pandas as pd
from collections import defaultdict
```

```python
scores = defaultdict(float)
with open(r'most_informative_features.txt') as f:
    for line in f.readlines():
        imp = line.split('\t')[2].split(',')
        for i in imp:
            i = i.strip().replace('"', '').replace("'", "")
            scores[i] += 1/6
```

```python
[x for x,y in sorted(scores.items(), key=lambda x: x[1], reverse=True) if y>0.3]
```
