```python
from __future__ import annotations
import networkx as nx
import netbiol3 as nb
a = nb.Abasy()

class GraphObserver:

    """A class to observe changes in a graph.
    
    TODO: To update. It doesn't check for changes in node/edges data yet.
    """

    def __init__(self, G: nx.Graph | nb.RegNet):
        """
        Initialize the GraphObserver class.
        
        Args:
            G (nx.Graph | nb.RegNet): The graph to observe.
        """
        self.graph = G
        self.current_edges = set(self.graph.edges())
        self.current_nodes = set(self.graph.nodes())
        self._changed = None

    def changed(self):
        """
        Check if the graph has changed since the last call.

        Returns:
            bool: True if the graph has changed, False otherwise.
        """
        new_edges = set(self.graph.edges())
        new_nodes = set(self.graph.nodes())

        if new_edges != self.current_edges or new_nodes != self.current_nodes:
            self._changed = True
            self.current_edges = new_edges
            self.current_nodes = new_nodes
        else:
            self._changed = False

        return self._changed


G = a.regnet('511145_v2003_sRDB01')
graph_observer = GraphObserver(G)
G.remove_node('crp')
graph_observer.changed()
```
