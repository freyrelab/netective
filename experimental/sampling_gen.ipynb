## Sampling Generator Fxn
#### Recibe
- Red como objeto RegNet
- Step (step para el porcentaje sampleo)
- Repeticiones por step
- Tipo de sampleo (nodos, edges, snowball)
#### Regresa
- Diccionario con llaves como el identificador y value el sampleo asociado

link_sampling - freyrelab.network.sampling


node_sampling - freyrelab.network.sampling

#### Imports

```python
from __future__ import annotations
from freyrelab.nets.sampling import edge_sampling, node_sampling, snowball_sampling
from freyrelab.regnets.regnet import RegNet
import networkx as nx
from netective.utils import *
```

#### Fxn

```python
def get_samples(G: RegNet | nx.DiGraph,  
                step: int=10, 
                rep: int=10,
                sampling_type : str='node' 
):
    """
        Function to create sampling networks.

        Args:
            G: RegNet or DiGraph.
                Network to compute the structural properties.
            step: int.
                Base level of network coverage for sampling and step for each
                increase until reaching 100% of coverage.
                Default is 10.
            rep: int.
                Number of samples per step.
                Default is 10.
            sampling_type: str.
                Network attribute by which sampling is done.
                Sampling types admitted: node-based, edge-based and snowball effect.
                Default is node-based sampling.

        Returns:
            nets : dictionary where keys as IDs for each network and its value is the sampled network.

        Raises:
            TypeError: If G is not a RegNet or DiGraph.
            ValueError: If user introduces a sampling type not admitted.
        """
    
    G = validate_network(G)
    # Validation for admitted sampling options
    sampling_opt = ['node', 'edge', 'snowball']
    if sampling_type not in sampling_opt:
        raise ValueError(f'Invalid sampling type, sampling options: {sampling_opt}')
    
    sampling_fxns = {
        'node' : node_sampling,
        'edge' : edge_sampling,
        'snowball' : snowball_sampling
    }
    
    nets = {
        f'Net_{step * x}_{j + 1}' : sampling_fxns[sampling_type](G, size=i)
        for x,i in enumerate(range(step, 100, step), 1)
        for j in range(rep)
    }
    
    return nets
```
