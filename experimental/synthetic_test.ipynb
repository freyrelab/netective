```python
import networkx as nx

from freyrelab.regnets.regnet import RegNet
from netective.pipeline import compare_networks
```

```python
networks = {}

for i in range (100, 1100, 100):
    networks[f'ER_{i}'] = nx.erdos_renyi_graph(i, 0.1)
    networks[f'BA_{i}'] = nx.barabasi_albert_graph(i, 10)
    networks[f'WS_{i}'] = nx.watts_strogatz_graph(i, 10, 0.1)
    networks[f'CG_{i}'] = nx.complete_graph(i)
    networks[f'SF_{i}'] = nx.scale_free_graph(i)
```

```python
compare_networks(networks)
```

```python
compare_networks(networks, norm='network')
```

```python
compare_networks(networks, norm='biological')
```
