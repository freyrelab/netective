# Use copy of a graph:

```python
class Foo:

    def __init__(self, G: nx.DiGraph):
        # self.G = G
        self.G = G.copy()
    
    def nodes_edges(self):
        print(f"In Foo G has {len(self.G.nodes)} nodes and {len(self.G.edges)} edges")
```

```python
G = nx.erdos_renyi_graph(100, 0.1)
G.add_edge(1, 1)
G.add_edge(2, 2)
print(f"G has {len(G.nodes)} nodes and {len(G.edges)} edges")

foo = Foo(G)
G.remove_edges_from(nx.selfloop_edges(G))

foo.nodes_edges()
```

# What is happening? ._.

```python
import networkx as nx
from freyrelab.regnets.regnet import RegNet

H = nx.erdos_renyi_graph(100, 0.1, directed=True, seed=1)
F = RegNet(H)

print(f'DiGraph --- H has {H.number_of_edges()} edges with number_of_edges and {len(H.edges())} with len(edges())')
print(f'RegNet --- F has {F.number_of_edges()} edges with number_of_edges and {len(F.edges())} with len(edges())')
print(f'self loops --- H has {nx.number_of_selfloops(H)} and F has {nx.number_of_selfloops(F)}')
print(f'difference in the set of edges --- {set(H.edges()) - set(F.edges())}')
```
